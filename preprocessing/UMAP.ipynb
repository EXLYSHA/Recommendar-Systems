```python
import numpy as np
import umap
import matplotlib.pyplot as plt
import os
from sklearn.decomposition import PCA
```

```python
os.chdir('/home/exlysha/MMRec/data/baby')
os.getcwd()
```

```python
file_path = './'
image_path = 'image_feat_raw.npy'
text_path = 'text_feat_raw.npy'
image_feat = np.load(os.path.join(file_path, image_path))
text_feat = np.load(os.path.join(file_path, text_path))
```

```python
# 2.  UMAP  2D
umap_model = umap.UMAP(n_components=2, random_state=42)

image_2d = umap_model.fit_transform(image_feat)
text_2d = umap_model.fit_transform(text_feat)

# 3. 
plt.figure(figsize=(12, 8))
for i in range(len(image_2d)):
    plt.plot([image_2d[i, 0], text_2d[i, 0]], [image_2d[i, 1], text_2d[i, 1]], 
             c='gray', linewidth=0.5, alpha=0.3)

# 
plt.scatter(image_2d[:, 0], image_2d[:, 1], c='blue', label='Image Features', s=10)

# 
plt.scatter(text_2d[:, 0], text_2d[:, 1], c='red', label='Text Features', s=10)

plt.legend()
plt.title('UMAP Projection of Image and Text Features')
plt.xlabel('UMAP-1')
plt.ylabel('UMAP-2')
plt.grid(True)
plt.tight_layout()
plt.show()
```
