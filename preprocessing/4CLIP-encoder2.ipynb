```python
path='downloaded_images'

currect=os.listdir(path)
currect=list(map(lambda x:x.split('.')[0],currect))
currect=list(map(int,currect))
currect.sort()
# currect[-1]
lost=[]
for i in range(0,currect[-1]+1):
    if i not in currect:
        lost.append(i)
lost
```

```python
import clip
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load('ViT-L/14@336px', device)
import numpy as np
vis_path = 'downloaded_images'
vis_list=os.listdir(vis_path)
vis_list.sort(key=lambda x: int(x.split('.')[0]))
vis_list=list(map(lambda x:os.path.join(vis_path,x),vis_list))
vis_list
```

```python
# batch = torch.cat([
#     preprocess(Image.open(path)).unsqueeze(0) 
#     for path in vis_list[:5]
# ], dim=0).to(device)
# with torch.no_grad():
#     image_features = model.encode_image(batch)
#     image_features /= image_features.norm(dim=-1, keepdim=True)
# image_features.shape
visual_embeddings=np.zeros((len(vis_list), 768))
for i in range(0, len(vis_list), 100):
    batch = torch.cat([
        preprocess(Image.open(path)).unsqueeze(0) 
        for path in vis_list[i:i+100]
    ], dim=0).to(device)
    with torch.no_grad():
        image_features = model.encode_image(batch)
        image_features /= image_features.norm(dim=-1, keepdim=True)
    visual_embeddings[i:i+100] = image_features.cpu().numpy()
    print(f'processed {i} to {i+100}')
visual_embeddings.shape
```

```python
assert visual_embeddings.shape[0] == len(vis_list)
np.save('vis_feat_vit_336.npy', visual_embeddings)
print('done!')
```
